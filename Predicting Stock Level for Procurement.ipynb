# Section 5 - Merge Data

In order to create the predictive model, we have to include all of the three datasets that we have. 

Remember the problem statement :

```
"Can we accurately predict the stock levels of products, based on sales data and sensor data, on an hourly basis in order to more intelligently procure products from out suppliers"
```

Based on the statement above, we know that we need to find which parameter will affect the predictive model significantly.

The client indicates that they want the model to predict on an hourly basis. By looking at every column, the only-column that exist in every dataset is 'timestamp' column.

In order to do that, we have to transform the timestamp column into hourly column for all of the datasets

````python
from datetime import datetime
def convert_timestamp_to_hourly(data : pd.DataFrame = None, column: str=None):
  # Duplicate the original data
  dummy = data.copy()
  # Convert array to list data
  new_ts = dummy[column].tolist()
  # Format the timestamp col as desired
  new_ts = [i.strftime('%Y-%m-%d %H:00:00') for i in new_ts]
  # Parse the timestamp in
  new_ts = [datetime.strptime(i, '%Y-%m-%d %H:00:00') for i in new_ts]
  dummy[column] = new_ts
  return dummy
````

````python
sales_df = convert_timestamp_to_hourly(sales_df, 'timestamp')
sales_df.head()
````

````python
stock_df = convert_timestamp_to_hourly(stock_df, 'timestamp')
stock_df.head()
````

````python
temp_df = convert_timestamp_to_hourly(temp_df, 'timestamp')
temp_df.head()
````

We can observe that the `timestamp` now has only hours, while the minutes and seconds are reduced to zero.

The next step is to aggregate each of column from the dataset which has the same value of `timestamp`. First, we aggregate the quantity.

For the `sales` data, we aggregate the sum of `quantity` column by grouping it based on `product ID` and `timestamp` column.

For the `stock` data, we aggregate the mean of `estimated_stock_pct` column by grouping it based on `product ID` and `timestamp` column.

For the `temp` data, we aggregate the mean of `temperature` column by grouping it based on `timestamp` column.

````python
sales_agg = sales_df.groupby(['timestamp','product_id']).agg({'quantity':'sum'}).reset_index()
sales_agg.head()
````

````python
stock_agg = stock_df.groupby(['timestamp','product_id']).agg({'estimated_stock_pct':'mean'}).reset_index()
stock_agg.head()
````

````python
temp_agg = temp_df.groupby(['timestamp']).agg({'temperature':'mean'}).reset_index()
temp_agg.head()
````

We will now merge the tables. We use `stock_agg` as our base table, which is merge with `sales_agg`

````python
merged_df = stock_agg.merge(sales_agg, on = ['timestamp','product_id'], how='left')
merged_df.head()
````

Now, we merge the `stock_agg` with `temp_agg`

````python
merged_df = merged_df.merge(temp_agg, on = 'timestamp')
#dff = pd.DataFrame(merged_df)
#dff.drop(['temperature_x'])
merged_df.head()
````

````python
merged_df.info()
````

# Section 7 - Modelling

We will use a supervised machine learning model. The target variable will be `estimated_stock_pct`, since the problem statement relies on how to manage the stock procurement on hourly basis.

To conduct cross-validation, we will use `K-Fold` method to train machine learning model on `K` (decided integer) random samples of the data.

First, we create the target variable `y` and independent variable `X`

````python
X = merged_df.drop(columns=['estimated_stock_pct'])
y = merged_df['estimated_stock_pct']
print(X.shape)
print(y.shape)
````

Based on the shape above, there is 28 predictor variables that will be trained on the machine learning model and 10845 rows of data.

Next, we define :
1. How many folds we want to complete during training
2. How many percentage of data that is going to be trained and tested

Usually, 20-30% of the data is used for testing.

````python
# Assign the K value
K = 10
split = 0.75
````

We are going to use `RandomForestRegressor` model, which is an instance of a Random Forest. These are powerful tree based ensemble algorithms and are particularly good because thei results are very interpretable.

For the machine learning algorythm, we will user `Regression`, because the `estimated_stock_pct` is a continuous numeric column. Meanwhile, the `classification` algorythm is conducted to predict a binary outcome.

The machine learning package would be `scikit-learn`, so we must import and install this package.

````python
!pip install scikit-learn
````

````python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import StandardScaler
````

Create a loop to train `K` models with a 75%/25% random split of the data, each time between training and test samples

````python
accuracy = []

for fold in range(0, K):

  # Define algorythm
  model = RandomForestRegressor()
  scaler = StandardScaler()

  # Create training & test samples
  X_train, X_test, y_train, y_test = train_test_split(X,y, train_size = split, random_state=42)


  # Scale X data, to converge the algorythm and to stay stabilized when handle big values
  scaler.fit(X_train)
  X_train=scaler.transform(X_train)
  X_test=scaler.transform(X_test)

  # Train model
  trained_model = model.fit(X_train, y_train)

  # Generate predictions on test sample
  y_pred = trained_model.predict(X_test)
   # Compute accuracy usin MAE (Mean Absolute Error)
  mae = mean_absolute_error(y_true = y_test, y_pred = y_pred)
  accuracy.append(mae)
  print(f'Fold {fold + 1}: MAE = {mae:.3f}')

  
print(f'Average MAE : {(sum(accuracy)/len(accuracy)) : .2f}')
````

Based on the iteration above, there are several important things to mention : 
1. The MAE remains the same for every iteration. This indicate a good consistency of the model across different random samples of the data. This is called Robust model

2. Even though it is robust, the value of MAE is still big, since we want the MAE to be as low as possible. The solution to this issue is to iteratively feature engineer the data, or adding more databases.
